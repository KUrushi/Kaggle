```python
from keras import layers,models
import matplotlib.pyplot as plt
from keras.utils import to_categorical
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('data/train.csv')
```

```python
data.head()
```

```python
data.drop(columns=['Name','Cabin','PassengerId','Ticket'],inplace=True)
data.head()
```

```python
data['Sex']=data['Sex'].apply(lambda x : 1 if x =='male' else 0)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data=data.drop(data[data['Embarked'].isnull()].index)
```

```python
data.shape
```

```python
mean_men=np.mean(data[(data['Sex']==1) &( False==(data['Age'].isnull()))]['Age'])
mean_women=np.mean(data[(data['Sex']==0) &( False==(data['Age'].isnull()))]['Age'])
```

```python
mean_men
```

```python
mean_women
```

```python
data[(data['Sex']==1) &((data['Age'].isnull()))]=data[(data['Sex']==1) &((data['Age'].isnull()))].fillna(mean_men)
```

```python
data[(data['Sex']==0) &((data['Age'].isnull()))]=data[(data['Sex']==0) &((data['Age'].isnull()))].fillna(mean_women)
```

```python
data.isnull().sum()
```

```python
np.unique(data['Parch'].values)
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['Embarked','Pclass','SibSp','Sex'])
data.head()
```

```python
data = pd.get_dummies(data, columns=['Parch'])
data.head()
```

```python
data['Fare']=data['Fare'].apply(lambda x : x/(np.max(data['Fare'])))
```

```python
data.head()
```

```python
model = models.Sequential()
model.add(layers.Dense(64,activation='relu',input_shape=(24,)))
model.add(layers.Dense(12,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer='rmsprop')
```

#### validation


```python
x_val=data.drop('Survived',axis=1)[:100].values
y_val=data.drop('Survived',axis=1)[:100].values
x_train=data.drop('Survived',axis=1)[100:].values
y_train=data.drop('Survived',axis=1)[100:].values
```

```python
history=model.fit(x_train,y_train,validation_data=(x_val,y_val),epochs=20,batch_size=128)
```
